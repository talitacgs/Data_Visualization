# Relplot

Relational plot são gráficos parecidos com scatterplot, porém com mais argumentos. A maior diferença é que também consegue [separar os atributos em linhas e colunas](https://www.futurelearn.com/info/courses/data-visualisation-with-python-seaborn-and-scatter-plots-sc/0/steps/192964), utilizando parametros col ou row, isso permite criar [subplots](https://medium.com/swlh/relplot-in-seaborn-69123bedcfe4) em uma mesma feature.

**Dataset**

```python
tips = sns.load_dataset("tips")

sns.relplot(x="total_bill", y="tip", data=tips);
plt.show()
```

**Visualizando a relação entre as variávels total da conta e gorjeta e colorindo cada ponto com a variável smoker(fumante)**

```python
# hue = parametro semantico para colorir o gráfico
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
plt.show()
```

**Visualizando a relação entre as variávels total da conta e gorjeta e colorindo cada ponto com a variável sex**

```python
sns.relplot(x="total_bill", y="tip", hue="sex", data=tips);
plt.show()
```

**Visualizando a relação entre as variávels total da conta e gorjeta e colorindo cada ponto com a variável  sexo e definindo um estilo também baseado na variável sexo**

```python
# Style = variavel semantica para definir estilo
sns.relplot(x="total_bill", y="tip", hue="sex", style="sex", data=tips);
plt.show()
```

**É possível representar quatro variáveis alterando a tonalidade e estilo de cada ponto. Mas é preciso cautela, pois, a visualização pode ficar confusa.**

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="smoker", data=tips);
plt.show()
```

**Em casos de variáveis numéricas para definir a tonalidade a paleta de cores altera de forma seguencial.**

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);
plt.show()
```

**Usando a opção size para definir o tamanho de cada ponto baseado em alguma coluna da base de dados**

```python
sns.relplot(x="total_bill", y="tip", size="size", data=tips);
plt.show()
```

```python
#sizes = parametro que define uma escala de tamanho para a representação do parametro size
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15,300), data=tips);
plt.show()
```

# Scatterplot

Scatterplot é um gráfico apensa de dispersão, possui muitos parâmetros iguais ao Relplot, porém não contém a função *lineplot()*. Com muitos dados, o scatterplot pode se tornar muito bagunçado para ter algum significado

**Dataset**

```python
df_iris = sns.load_dataset("iris")
df_iris.head(5)
```

**Plot da relação entre a largura da sépala com o comprimento da pétala**

```python
fig = plt.figure(figsize=(15,5))

sns.scatterplot(data=df_iris, x="sepal_width", y="petal_length")
plt.show()
```

**Plot da relação entre o comprimento da sépala com o comprimento da pétala**

```python
sns.scatterplot(data=df_iris, x="sepal_length", y="petal_length")
plt.show()
```

**Plot da relação entre a largura da sépala com o comprimento da pétala distinguindo pelas espécies utilizando o argumento hue**

```python
sns.scatterplot(data=df_iris, x="sepal_width", y="petal_length",hue="species")
plt.show()
```

## Desafio

**Plote um scatter plot para cada combinação entre as colunas numéricas utilizando o argumento hue para diferenciar as espécies (sepal_length, sepal_width, petal_length, petal_width)**

```python
#                                   Apenas colunas numéricas
for x_iterator in df_iris.select_dtypes(include=np.number):
  for y_iterator in df_iris.select_dtypes(include=np.number):

    if x_iterator != y_iterator:

      sns.scatterplot(data=df_iris, x=x_iterator, y=y_iterator, hue="species")
      plt.show()
```

# Jointplot

**Plot da relação entre o comprimento da sépala com o comprimento da pétala**

```python
sns.jointplot(data=df_iris, x="sepal_length", y="petal_length")
plt.show()
```

**Plot da relação entre o comprimento da sépala com o comprimento da pétala distinguindo pelas espécies utilizando o argumento hue**

**obs: Quando o argumento hue é utilizado ao invés de plotar histogramas junto com o scatter plot é exibido as curvas KDE**

```python
sns.jointplot(data=df_iris, x="sepal_length", y="petal_length",hue="species")
plt.show()
```

# Pairplot

**Este método cria gráficos de dispersão de todas as combinações de variáveis numéricas dentro do DataFrame e histogramas na diagonal.**

```python
sns.pairplot(df_iris)
plt.show()
```
