**Biblioteca**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

**Dataset**

```python
tips = sns.load_dataset("tips")
```

# Boxplot

**Boxplot de Dias da semana por total de conta**

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips);
plt.show()
```

**Gráfico de Boxplot dos dias por total de conta e se a pessoa é fumante ou não.**

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips);
plt.show()
```

## Boxen

* Tipo de Boxplot com foco maior na forma dos dados do que nos outliers.

```python
sns.catplot(x="day", y="total_bill", kind="boxen", data=tips);
plt.show()
```

**Combinando com o método stripplot**

```python
sns.catplot(x="day", y="total_bill", kind="boxen", data=tips);
sns.stripplot(x="day", y="total_bill", data=tips, color="blue");
plt.show()
```

```python
sns.catplot(x="sex", y="total_bill", hue="smoker", col="time", data=tips, kind="boxen",height=4);
plt.show()
```

**Dataset**

```python
df_iris = sns.load_dataset("iris")
df_iris.head(5)
```

**Box-Plot do comprimento da pétala**

```python
sns.boxplot(data=df_iris, x="species", y="petal_length")
plt.show()
```

**Box-Plot do comprimento da pétala sem os outliers**

```python
sns.boxplot(data=df_iris, x="species", y="petal_length", showfliers = False)
plt.show()
```

**Box-Plot de todas as variáveis numéricas**

```python
for coluna in df_iris.select_dtypes(include=np.number):

  sns.boxplot(data=df_iris, x="species", y=coluna)
  plt.show()
```
