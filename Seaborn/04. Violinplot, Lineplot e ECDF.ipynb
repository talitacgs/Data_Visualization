**Biblioteca**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

**Dataset**

```python
tips = sns.load_dataset("tips")
```

# Violinplot

```python
sns.catplot(x="day", y="total_bill", kind="violin", data=tips);
plt.show()
```

```python
sns.catplot(x="total_bill", y="day", kind="violin", data=tips);
plt.show()
```

**Usando o parâmetro Hue para ter uma terceira variável**

```python
sns.catplot(x="day", y="total_bill", hue="time", kind="violin", data=tips);
plt.show()
```

**Utilizando o método split**

```python
sns.catplot(x="day", y="total_bill", hue="sex", split=True , kind="violin", data=tips);
plt.show()
```

**Usando o parametro inner para preencher o violin plot com observações**

```python
sns.catplot(x="day", y="total_bill", hue="sex", split=True , inner="stick", palette="pastel", kind="violin", data=tips);
plt.show()
```

**Combinando gráfico de swarm com o Violin Plot**

```python
sns.catplot(x="day", y="total_bill", palette="pastel", kind="violin", data=tips);
sns.swarmplot(x="day", y="total_bill", color="k", size=3, data=tips);
plt.show()
```

**Dataset**

```python
df_iris = sns.load_dataset("iris")
```

**Violin Plot do comprimento da pétala**

```python
sns.violinplot(data=df_iris, x="species", y="petal_length");
plt.show()
```

**Violin Plot da largura da pétala**

```python
sns.violinplot(data=df_iris, x="species", y="petal_width");
plt.show()
```

**Violin Plot de todas as variáveis numéricas**

```python
for coluna in df_iris.select_dtypes(include=np.number):

  sns.violinplot(data=df_iris, x="species", y=coluna)
  plt.show()
```

# Lineplot

**Dataset**

```python
flights = sns.load_dataset("flights")
```

```python
sns.lineplot(data=flights, x="year", y="passengers")
plt.show()
```

**Plot da relação entre o número de passageiros e os anos distinguindo pelos meses utilizando o argumento hue**

```python
fig= plt.figure(figsize=(10,5))
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
plt.show()
```

# ECDF

**ECDF do comprimento da pétala**

```python
sns.ecdfplot(data=df_iris, x="petal_length");
plt.show()
```

**ECDF do comprimento da sépala**

```python
sns.ecdfplot(data=df_iris, x="sepal_length");
plt.show()
```

**ECDF do comprimento da pétala utilizando o argumento hue**

```python
sns.ecdfplot(data=df_iris, x="sepal_length", hue="species");
plt.show()
```

**ECDF de todas as variáveis numéricas**

```python
for coluna in df_iris.select_dtypes(include=np.number):
  sns.ecdfplot(data=df_iris, hue="species", x=coluna)
  plt.show()
```
