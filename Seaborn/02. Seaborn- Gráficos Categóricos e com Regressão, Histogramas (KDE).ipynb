**Biblioteca**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

**Dataset**

```python
tips = sns.load_dataset("tips")
tips.head()
```

# Gráficos Categóricos

**Gráfico de barras**

```python
sns.catplot(x="sex", kind="count", palette="Set2", data=tips);
plt.show()
```

**Gráfico de barras horizontais usando a coluna *day***

```python
sns.catplot(y="day", kind="count", palette="Set1", data=tips);
plt.show()
```

**Scatter plot com dados categóricos**

```python
sns.catplot(x="day", y="total_bill", palette="Set2", data=tips);
plt.show()
```

```python
# O parâmetro swarm evita sobreposição de pontos
sns.catplot(x="day", y="total_bill", kind='swarm', data=tips);
plt.show()
```

**O parâmetro hue permite adicionarmos uma terceira variável a nossa visualização**

```python
sns.catplot(x="day", y="total_bill", kind='swarm', hue="sex", palette="Dark2", data=tips);
plt.show()
```

```python
# O parametro **order** permite alterarmos a ordem padrão das categorias que estão sendo exibidas.
sns.catplot(x="day", y="total_bill", kind='swarm', hue="sex", order=['Sat','Sun','Fri','Thur'], data=tips);
plt.show()
```

**Invertendo a visualização para plots horizontais**

```python
sns.catplot(x="total_bill", y="day", kind="swarm", data=tips);
plt.show()
```

# Histograma (KDE)

```python
df_iris = sns.load_dataset("iris")
df_iris.head(5)
```

**Distribuição da largura da pétala**

```python
sns.displot(data=df_iris, x="petal_width",bins=10)
plt.show()
```

**Distribuição da largura da pétala para cada espécie uma do lado da outra (utilizando o argumento col)**

```python
sns.displot(data=df_iris, x="petal_width", col="species")
plt.show()
```

**Distribuição da largura da pétala para cada espécie uma embaixo da outra (utilizando o argumento row)**

```python
sns.displot(data=df_iris, x="petal_width", row="species")
plt.show()
```

## Desafio 

**Plote um histograma para cada coluna numérica (sepal_length, sepal_width, petal_length, petal_width)**

```python
for coluna in df_iris.select_dtypes(include=np.number):

  sns.displot(data=df_iris, x=coluna, col="species")
  plt.show()
```

## Displot com KDE = True

**A função displot utilizando o argumento KDE = True cria automaticamente a curva *KDE (Kernel Density Estimation)*, a qual consiste em uma *estimativa para a função densidade de probabilidade***

```python
for coluna in df_iris.select_dtypes(include=np.number):

  sns.displot(data=df_iris, x=coluna, col="species",kde=True)
  plt.show()
```

O parâmetro hue se utiliza dos valores categóricos de uma coluna ou lista de valores para serem diferenciadas por suas cores na visualização.

```python
for coluna in df_iris.select_dtypes(include=np.number):

  sns.displot(data=df_iris, x=coluna, hue="species", kind="kde", palette="Accent")
  plt.show()
```
