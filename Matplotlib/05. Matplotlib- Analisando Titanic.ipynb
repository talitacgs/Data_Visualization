```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

**Relação entre a Idade (age) e o preço pago (fare) para embarcar**

```python
#plot(x,y)
plt.scatter(x="age",y="fare",data=titanic,color="#041562")

#Definindo labels
plt.xlabel("age")
plt.ylabel("fare")

plt.show()
```

**Relação entre a Idade (age) e o preço pago (fare) para cada sexo sobreposto**

```python
#plot(x,y)
plt.scatter(x="age",y="fare",data=titanic[titanic["sex"]=="male"],label="male",color="#041562")
plt.scatter(x="age",y="fare",data=titanic[titanic["sex"]=="female"],label="female",color="#DA1212")

#Definindo labels
plt.xlabel("age")
plt.ylabel("fare")

#Legendas
plt.legend()

plt.show()
```

**Histograma da idade (age) para cada sexo sobreposto**

```python
#plot(x,y)
plt.hist(x="age",data=titanic[titanic["sex"]=="male"],label="male",color="#041562")
plt.hist(x="age",data=titanic[titanic["sex"]=="female"],label="female",color="#DA1212")

#Definindo labels
plt.xlabel("age")
plt.show()
```

## Desafio 1

**Plote um scatter plot entre a Idade (age) e o preço pago (fare) para sobreviventes e não sobreviventes sobrepostos**

**Plote um Histograma plot da Idade (age) para sobreviventes e não sobreviventes sobrepostos**

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(15,5))

#Scatter plot
#plot(x,y)
axes[0].scatter(x="age",y="fare",data=titanic[titanic["survived"]==0],label="non-survived",color="#11468F")
axes[0].scatter(x="age",y="fare",data=titanic[titanic["survived"]==1],label="survived",color="#00ABB3")

#Definindo labels
axes[0].set(xlabel="age",
            ylabel="fare",
           title="Idade e preço pago de sobreviventes e nao sobreviventes")

#Exibir legenda
axes[0].legend()

#Histogram
#plot(x,y)
axes[1].hist(x="age",data=titanic[titanic["survived"]==0],label="non-survived",color="#11468F")
axes[1].hist(x="age",data=titanic[titanic["survived"]==1],label="survived",color="#00ABB3",alpha=0.3)

#Definindo labels
axes[1].set(xlabel="age",
            ylabel="fare",
           title="Idade de sobreviventes e nao sobreviventes")

#Exibir legenda
axes[1].legend()

plt.show()
```

**BoxPlot do preço pago (fare)**

```python
c= '#041562'
c2= "#11468F"
c3= "#DA1212"

#plot(x,y)
plt.boxplot(x=titanic["fare"],
            vert=False,
            patch_artist=True,    # ativa preenchimento com cor
            boxprops=dict(facecolor=c, color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c2, markeredgecolor=c2),
            medianprops=dict(color=c3))

#Definindo label
plt.xlabel("fare")

plt.show()
```

**BoxPlot da idade (age)**

```python
c= '#041562'
c2 = "#DA1212"

#plot(x,y)
plt.boxplot(x=titanic.dropna()["age"],
            vert=False,           # alinhamento vertical
            patch_artist=True,    # ativa preenchimento com cor
            boxprops=dict(facecolor=c, color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color=c2))

#Definindo label
plt.xlabel("age")

plt.show()
```

**BoxPlot da idade (age) para sobreviventes e não sobreviventes**

```python
survivor = titanic[titanic["alive"]=="yes"].dropna()["age"]
non_survivor = titanic[titanic["alive"]=="no"].dropna()["age"]

#plot(x,y)
bplot1=plt.boxplot(x=[survivor,non_survivor],
            patch_artist=True,
            boxprops=dict(facecolor=c, color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color=c2))

plt.xticks([1, 2], ['survivor', 'non-survivor'])

#Definindo labels
plt.ylabel("age")

plt.show()
```

**Barplot vertical da quantidade de passageiros por cidade de embarque**

```python
titanic_grouped = titanic.groupby("embark_town").count()

altura = []
for i in titanic_grouped["alive"]:
   altura.append(i)
posicao = []
for i in range(0,3,1): 
   posicao.append(i)

#plot(x,y)
plt.figure(figsize=(8, 8))
plt.bar(x=titanic_grouped.index,height=titanic_grouped["alive"].values,color=c)

#Definindo labels
plt.xlabel("embark_town")
plt.ylabel("count passangers")

#Colocando a quantidade em cada barra - x=posição, y=altura, s=altura
for i in range(3):
   plt.text(x=posicao[i]-0.15, y=altura[i]+5, s=str(altura[i]),   
           fontsize=15)

plt.show()
```

**Barplot vertical da quantidade de passageiros por cidade de embarque empilhados por sobreviventes e não sobreviventes**

```python
titanic_grouped_survivor = titanic[titanic["alive"]=="yes"].groupby("embark_town").count()
titanic_grouped_non_survivor = titanic[titanic["alive"]=="no"].groupby("embark_town").count()

altura1 = []
for i in titanic_grouped_survivor["alive"]:
   altura1.append(i)
altura2 = []
for i in titanic_grouped_non_survivor["alive"]:
   altura2.append(i)
posicao1 = []
for i in range(0,3,1): 
   posicao.append(i)

#plot(x,y)
plt.figure(figsize=(8, 8))
plt.bar(x=titanic_grouped_survivor.index,
        height=titanic_grouped_survivor["alive"].values,label="survivor",color="#00ABB3")

plt.bar(x=titanic_grouped_non_survivor.index,
        height=titanic_grouped_non_survivor["alive"].values,
        bottom= titanic_grouped_survivor["alive"].values, #Indicar onde deve começar o gráfico a ser empilhado
        label="non-survivor",
        color="#11468F")

#Colocando a quantidade em cada barra - x=posição, y=altura, s=altura
for i in range(3):
   plt.text(x=posicao[i]-0.15, y=altura1[i]+10, s=str(altura1[i]),   
           fontsize=15)

#Colocando a quantidade em cada barra - x=posição, y=altura, s=altura
for i in range(3):
   plt.text(x=posicao[i]-0.15, y=titanic_grouped_survivor["alive"].values[i] + altura2[i] + 10, s=str(altura2[i]),   
           fontsize=15)

#Inserir legenda
plt.legend()

plt.show()
```

**LinePlot da quantidade de passageiros por cidade de embarque para cada classe**

```python
titanic_grouped_class_1 = titanic[titanic["class"]=="First"].groupby("embark_town").count()
titanic_grouped_class_2 = titanic[titanic["class"]=="Second"].groupby("embark_town").count()
titanic_grouped_class_3 = titanic[titanic["class"]=="Third"].groupby("embark_town").count()

#plot(x,y)
plt.plot(titanic_grouped_class_1.index,
        titanic_grouped_class_1['alive'].values,
        label="First",
        color="#041562")

plt.plot(titanic_grouped_class_2.index,
        titanic_grouped_class_2['alive'].values,
        label="Second",
        color="#11468F",
        ls="--")

plt.plot(titanic_grouped_class_3.index,
        titanic_grouped_class_3['alive'].values,
        label="Third",
        color="#DA1212")

# Inserir legenda
plt.legend()

plt.show()
```

**Gráfico de pizza da quantidade de passageiros separado por classes com fatias destacadas**

```python
titanic_grouped = titanic.groupby("class").count()

plt.pie(x=titanic_grouped["alive"].values,
        labels=titanic_grouped.index,
        explode=[0.2,0.2,0.1],
        autopct='%1.1f%%',
        colors=["#041562","#11468F","#DA1212"])

plt.show()
```

## Desafio 2

**LinePlot da quantidade de passageiros por classe para sobrevivente e não sobrevivente**

**Barplot vertical da quantidade de passageiros por cidade de embarque com as barras empilhadas por classe**

**Scatter plot entre a Idade (age) e o preço pago (fare) para cada classe**

**Gráfico de pizza da quantidade de passageiros separado por cidade de embarque**

```python
fig,axes= plt.subplots(2,2,figsize=(20,18))

#lineplot
titanic_grouped_survivor = titanic[titanic["alive"]=="yes"].groupby("class").count()
titanic_grouped_non_survivor = titanic[titanic["alive"]=="no"].groupby("class").count()

#plot(x,y)
axes[0,0].plot(titanic_grouped_survivor.index,
               titanic_grouped_survivor["alive"].values,
               label= "survivor",
               color= "#00ABB3")
axes[0,0].plot(titanic_grouped_non_survivor.index,
               titanic_grouped_non_survivor["alive"].values,
               label= "non survivor",
               color= "#11468F")

#Inserindo legenda
axes[0,0].legend()

#Inserindo labels e título
axes[0,0].set(xlabel="class", 
              ylabel="count passengers", 
              title="LinePlot da quantidade de passageiros por classe para sobrevivente e não sobrevivente")
#-----------------------------------------------------------------------------------------------------------------------------#
#Barplot
titanic_grouped_class_1 = titanic[titanic["class"]== "First"].groupby("embark_town").count()
titanic_grouped_class_2 = titanic[titanic["class"]== "Second"].groupby("embark_town").count()
titanic_grouped_class_3 = titanic[titanic["class"]== "Third"].groupby("embark_town").count()

#plot(x,y)
axes[0,1].bar(titanic_grouped_class_1.index,
             titanic_grouped_class_1["alive"].values,
             label="First",
             color="#041562")
axes[0,1].bar(titanic_grouped_class_2.index,
             titanic_grouped_class_2["alive"].values,
             label="Second",
             color="#11468F",
             bottom=titanic_grouped_class_1["alive"].values)
axes[0,1].bar(titanic_grouped_class_3.index,
             titanic_grouped_class_3["alive"].values,
             label="Third",
             color="#DA1212",
             bottom=titanic_grouped_class_2["alive"].values + titanic_grouped_class_1["alive"].values)

#Inserir legenda
axes[0,1].legend()

#Inserindo labels e título
axes[0,1].set(xlabel="Embarked Town",
              ylabel="count passengers",
              title="Barplot de passageiros por cidades de embarque agrupados por classe")

#-----------------------------------------------------------------------------------------------------------------------------#
#Scatter plot
#plot(x,y)
axes[1,0].scatter(x="age",y="fare",data=titanic[titanic["class"]=="First"],label="First Class",color="#041562")
axes[1,0].scatter(x="age",y="fare",data=titanic[titanic["class"]=="Second"],label="Second Class",color="#11468F")
axes[1,0].scatter(x="age",y="fare",data=titanic[titanic["class"]=="Third"],label="Third Class",color="#DA1212")

#Inserindo labels e título
axes[1,0].set(xlabel="age",
              ylabel="fare",
              title="Idade e preço pago no bilhete por classe")

#Inserir legenda
axes[1,0].legend()

#-----------------------------------------------------------------------------------------------------------------------------#
#Pie plot
titanic_embarked_town = titanic.groupby("embark_town").count()
labels = ["Queenstown","Cherbourg","Southampton"]

#plot(x,y)
axes[1,1].pie(x=titanic_embarked_town["alive"].values,
        labels=titanic_embarked_town.index,
        explode=[0.1,0.1,0.05],
        autopct='%1.1f%%',
        colors=["#041562","#11468F","#DA1212"],
        textprops={'color':"w",'fontsize': 16, "weight":"bold"}) #Mudar a cor da % do gráfico, tamanho e "peso"

#Inserindo labels e título
axes[1,1].set(title="Passageiros vs Cidades de embarque")

axes[1,1].legend(labels,
                 title="Embarked Town",
                 bbox_to_anchor=(-0.1, 0.8, 0.8, 0.1))

fig.show()
```
