**Bibliotecas**

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

**Dataframe**

```python
speed = [17.5, 40, 48, 52, 69, 88,129,143,97,64,46]
lifespan = [8, 70, 1.5, 25, 12, 28,80,32,35,24,20]
animals = ['pig', 'elephant',
         'rabbit', 'giraffe', 'coyote', 'horse', 'eagle','goose','pigeon','owl','sparrow']

type_animal = ["mammal","mammal","mammal","mammal","mammal","mammal",
           "bird","bird","bird","bird","bird"]

df_animals = pd.DataFrame({"speed":speed,
                           "lifespan":lifespan,
                           "animals":animals,
                           "type_animal": type_animal})

df_animals
```

# Scatterplot

```python
df_animals.plot.scatter(x='speed',y='lifespan')
plt.show()
```

**Diferenciando pássaros de mamíferos**

```python
df_animals.replace({"mammal":0,"bird":1}).plot.scatter(x='speed',
                                                              y='lifespan',
                                                              c='type_animal',colormap='viridis')
plt.show()
```

**Aumentando o tamanho dos pontos baseado no tempo de vida dos animais**

```python
df_animals.replace({"mammal":0,"bird":1}).plot.scatter(x='speed',
                                                       y='lifespan',
                                                       c='type_animal',
                                                       colormap='viridis',
                                                       s="lifespan")
plt.show()
```

# Boxplot

```python
df_animals.boxplot(column=['speed'])
plt.show()
```

**Espécies separados por velocidades**

```python
df_animals.boxplot(column='speed',by='type_animal',figsize=(7,7))
plt.show()
```

**Retirando grid**

```python
df_animals.boxplot(column='lifespan',by='type_animal',figsize=(7,7),grid=False);
plt.show()
```
