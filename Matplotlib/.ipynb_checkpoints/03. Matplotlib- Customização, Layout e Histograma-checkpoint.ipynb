**Bibliotecas**

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

# Customização

Para outros estilos: [Style Sheets](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)

```python
# Define o stilo para ggplot
plt.style.use("ggplot")
```

```python
#Opções do método axis
%matplotlib inline

# 100 valores no intervalo de 0 a 2.
x = np.linspace(0, 2, num=100)

# plot(x,y)
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratico', linewidth=5)

# Definindo label dos eixos
plt.xlabel('x label')
plt.ylabel('y label')

# Título
plt.title("Gráfico de Linhas Simples")

# Exibe a legenda e por padrão usa o label de cada plot.
plt.legend()

# Configurações do texto
plt.text(1.00, 1.0, "Cruzamento das Linhas", fontsize=12, horizontalalignment='right')

plt.show()
```

# Histograma

```python
speed = [17.5, 40, 48, 52, 69, 88,129,143,97,64,46]
lifespan = [8, 70, 1.5, 25, 12, 28,80,32,35,24,20]
animals = ['pig', 'elephant',
         'rabbit', 'giraffe', 'coyote', 'horse', 'eagle','goose','pigeon','owl','sparrow']

type_animal = ["mammal","mammal","mammal","mammal","mammal","mammal",
           "bird","bird","bird","bird","bird"]

df_animals = pd.DataFrame({"speed":speed,
                           "lifespan":lifespan,
                           "animals":animals,
                           "type_animal": type_animal})

df_animals
```

```python
df_animals.hist(column=["speed"],bins=5,grid=False);
plt.show()
```

**Alterando o número de bins**

```python
df_animals.hist(column=["speed"],bins=10,grid=False);
plt.show()
```

**Subplot**

```python
df_animals.hist(column=["speed","lifespan"])
plt.show()
```

**Remoção de grid**

```python
df_animals.hist(column=["speed","lifespan"],grid=False)
plt.show()
```

**Histograma da velocidade para cada tipo de animal (type_animal)**

```python
df_animals.hist(column="speed",bins=10,by="type_animal",figsize=(10,5))
plt.show()
```

**Histograma do tempo de vida para cada tipo de animal (type_animal)**

```python
df_animals.hist(column="lifespan",bins=10,by="type_animal",figsize=(10,5))
plt.show()
```
