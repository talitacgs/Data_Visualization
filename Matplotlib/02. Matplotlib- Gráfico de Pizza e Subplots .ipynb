# Gr√°fico de Pizza

```python
#Op√ß√µes do m√©todo axis
%matplotlib inline

# Dados
vendas = [3000, 2300, 1000, 500]
labels = ['E-commerce', 'Loja F√≠sica', 'e-mail', 'Marketplace']

#plot (x,y)
plt.pie(vendas, labels=labels)

plt.show()
```

**Adicionando informa√ß√µes de percentual, sombra, separa√ß√£o de partes e legenda**

```python
#Op√ß√µes do m√©todo axis
%matplotlib inline

#Dados
vendas = [3000, 2300, 1000, 500]
labels = ['E-commerce', 'Loja F√≠sica', 'e-mail', 'Marketplace']

# define o n√≠vel de separabilidade entre as partes, ordem do vetor representa as partes
explode = (0.1, 0, 0, 0) 

#plot(x,y)                     formata√ß√£o da %   
plt.pie(vendas, labels=labels, autopct='%1.1f%%', shadow=True, explode=explode, colors=["red","pink","blue","purple"])

#Legenda        localiza√ß√£o da legenda(existe op√ß√£o "best")
plt.legend(labels, loc=2)

# define que o gr√°fico ser√° plotado em circulo
plt.axis('equal')

plt.show()
```

**Acessando subplots com o m√©todo *subplot***

```python
#Dados
x = [1,2,3,4]
y = [10,20,25,40]

#plot_1(x,y) qtde_linha, qtde_coluna, acessando 1o gr√°fico
plt.subplot(1,2,1)
plt.plot(x,y, 'o')

#plot_2(x,y) qtde_linha, qtde_coluna, acessando 2o gr√°fico
plt.subplot(1,2,2)
plt.plot(x,y, color='r')
```

**Cria√ß√£o de inst√¢ncias de classes figure (menos c√≥digo devido a POO)**

```python
# Cria um objeto do tipo figure
fig = plt.figure()

# Define o tamanho em percentual do grafico.
# Altura e largura do eixo na figura
axes_a = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Plot(x,y)
axes_a.plot(x,y)
```

**Uma figura com dois gr√°ficos de propor√ß√µes diferentes**

```python
# Cria um objeto do tipo figure
fig = plt.figure()

# Define o tamanho em percentual do grafico.
# Altura e largura do eixo na figura
axes_a = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes_b = fig.add_axes([0.2, 0.5, 0.3, 0.3])

# Plota a figura com dois eixos.
# O "r--" significa linha tracejada em vermelho(r)
axes_a.plot(x,y)
axes_b.plot(x,y,'r--')
```

```python
# Cria um objeto do tipo figure
fig = plt.figure()

# left(esquerda), bottom(inferior), largura, altura do eixo na figura
axes_a = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes_b = fig.add_axes([0.8, 0.3, 0.3, 0.3])

# Plota a figura com dois eixos.
# O "r--" significa linha tracejada em vermelho(r)
axes_a.plot(x,y)
axes_b.plot(x,y,'r--')
```

**Defini√ß√£o dos eixos com o m√©todo *set()***

```python
#Op√ß√µes do m√©todo axis
import matplotlib.pyplot as plt

# Define as configura√ß√µes dos plots
# Cada plot ter√° o mesmo tamanho de figuras
# Ver nota
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

# Dados para cada subplot
ax1.bar([1,2,3],[3,4,5], color='r')
ax2.barh([0.5,1,2.5],[0,1,2])

# Usa o m√©todo set para definir configura√ß√µes do  gr√°fico
ax1.set(title="Gr√°fico de Barras Verticais", xlabel="Eixo x", ylabel=" Eixo y")
ax2.set(title="Gr√°fico de Barras Horizontais", xlabel="Eixo x", ylabel="Eixo y")

plt.show()
```

```python
# NOTA: Ao utilizar a orienta√ß√£o a objetos, em uma linha de c√≥digo definimos o tamanho de todos os gr√°ficos(figsize),
# utilizamos menos linha de c√≥digo, pois para cada gr√°fico, ter√≠amos que utilizar o m√©todo .subplot()
# conseguimos maior customiza√ß√£o ao acessar cada gr√°fico com o m√©todo .set()
```

**Subplots de mais de dois gr√°ficos**

```python
# Define as configura√ß√µes dos plots
# Cada plot ter√° o mesmo tamanho de figuras
fig, axs = plt.subplots(2, 2, figsize=(10,12)) # Cria 4 axes


# Dados para cada subplot
ax1 = axs[0, 0] #acessa cada subplot como uma matriz
ax1.bar([1,2,3],[3,4,5], color='r')

ax2 = axs[0, 1]
ax2.barh([0.5,1,2.5],[0,1,2])

ax3 = axs[1, 0]
ax3.scatter([0.5,1,2.5],[0,1,2], color='b')

ax4 = axs[1, 1]
ax4.plot([1,2,3],[3,4,5], '*', color = 'purple')


# Usa o m√©todo set para definir configura√ß√µes do gr√°fico
ax1.set(title="Gr√°fico de Barras Verticais", xlabel="Eixo x", ylabel=" Eixo y")
ax2.set(title="Gr√°fico de Barras Horizontais", xlabel="Eixo x", ylabel="Eixo y")
ax3.set(title="Gr√°fico Scatter", xlabel="Eixo x", ylabel=" Eixo y")
ax4.set(title="Gr√°fico de Estrelas", xlabel="Eixo x", ylabel="Eixo y")

plt.show()
```
